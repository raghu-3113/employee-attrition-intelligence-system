## Exploratory Data Analysis (EDA)

### Objective
The objective of this notebook is to perform exploratory data analysis on the employee attrition dataset to identify:
- Workforce patterns
- Attrition trends
- Employee behavior insights
- Relationships between important variables
- Key factors contributing to employee attrition

### Key Goals
- Analyze employee attrition distribution
- Study department-wise attrition trends
- Examine salary and experience patterns
- Investigate overtime and work-life balance impact
- Generate business-driven insights for workforce retention

```python
# ============================================
# Import Required Libraries
# ============================================

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

# Plot styling
plt.figure(figsize=(8, 5))

# Display settings
pd.set_option('display.max_columns', None)

print("Libraries imported successfully.")
```

# Load Cleaned Dataset

In this section, the cleaned employee attrition dataset is loaded for exploratory analysis.

```python
# ============================================
# Load Cleaned Dataset
# ============================================

df = pd.read_csv("../data/cleaned_employee_attrition.csv")

print("Dataset loaded successfully.")
```

```python
# Display first 5 rows
df.head()
```

# Attrition Distribution Analysis

Understanding the overall distribution of employee attrition helps identify:
- Workforce stability
- Retention concerns
- Dataset imbalance
- Scale of employee turnover

```python
# ============================================
# Attrition Distribution
# ============================================

plt.figure(figsize=(6,4))

sns.countplot(x='Attrition', data=df)

plt.title("Employee Attrition Distribution")
plt.xlabel("Attrition")
plt.ylabel("Employee Count")

plt.show()
```

### Insight

The attrition distribution reveals the proportion of employees who left the organization compared to those who remained.

A significant imbalance between classes may affect machine learning model performance and may require balancing techniques such as SMOTE during model training.

# Department-Wise Attrition Analysis

This analysis helps identify departments experiencing higher employee turnover.

Understanding department-level attrition is critical for:
- Workforce planning
- Resource allocation
- Department-specific retention strategies

```python
# ============================================
# Department-wise Attrition
# ============================================

plt.figure(figsize=(8,5))

sns.countplot(x='Department', hue='Attrition', data=df)

plt.title("Department-wise Employee Attrition")
plt.xlabel("Department")
plt.ylabel("Employee Count")

plt.xticks(rotation=15)

plt.show()
```

### Insight

The visualization highlights departments with comparatively higher attrition rates.

Departments with elevated employee turnover may require:
- Better employee engagement
- Improved workload management
- Career growth opportunities
- Leadership interventions

# Overtime Impact on Employee Attrition

Workload imbalance is often a major contributor to employee turnover.

This analysis evaluates whether overtime work is associated with higher attrition risk.

```python
# ============================================
# Overtime vs Attrition
# ============================================

plt.figure(figsize=(6,4))

sns.countplot(x='OverTime', hue='Attrition', data=df)

plt.title("Overtime vs Employee Attrition")
plt.xlabel("OverTime")
plt.ylabel("Employee Count")

plt.show()
```

### Insight

Employees working overtime appear to exhibit higher attrition rates compared to employees without overtime responsibilities.

This suggests that excessive workload and work-life imbalance may significantly contribute to employee dissatisfaction and turnover.

# Monthly Income Distribution

Salary plays an important role in employee satisfaction and retention.

This analysis examines the distribution of employee monthly income across the organization.

```python
# ============================================
# Monthly Income Distribution
# ============================================

plt.figure(figsize=(8,5))

sns.histplot(df['MonthlyIncome'], kde=True)

plt.title("Monthly Income Distribution")
plt.xlabel("Monthly Income")
plt.ylabel("Frequency")

plt.show()
```

### Insight

The salary distribution indicates the spread of employee compensation levels across the organization.

Large variations in salary distribution may influence:
- Employee satisfaction
- Motivation
- Retention probability

# Job Satisfaction vs Attrition

Job satisfaction is one of the most important factors influencing employee retention.

This analysis investigates whether lower job satisfaction levels correlate with higher attrition.

```python
# ============================================
# Job Satisfaction vs Attrition
# ============================================

plt.figure(figsize=(8,5))

sns.countplot(x='JobSatisfaction', hue='Attrition', data=df)

plt.title("Job Satisfaction vs Attrition")
plt.xlabel("Job Satisfaction")
plt.ylabel("Employee Count")

plt.show()
```

### Insight

Employees with lower job satisfaction levels tend to exhibit higher attrition tendencies.

Improving workplace engagement, recognition, and employee experience may help reduce attrition risk.

# Work-Life Balance Analysis

Maintaining a healthy work-life balance is critical for long-term employee retention and well-being.

This analysis examines the relationship between work-life balance ratings and attrition.

```python
# ============================================
# Work-Life Balance vs Attrition
# ============================================

plt.figure(figsize=(8,5))

sns.countplot(x='WorkLifeBalance', hue='Attrition', data=df)

plt.title("Work-Life Balance vs Attrition")
plt.xlabel("Work-Life Balance")
plt.ylabel("Employee Count")

plt.show()
```

### Insight

Employees reporting lower work-life balance scores show comparatively higher attrition levels.

This suggests that employee well-being initiatives may positively impact workforce retention.

# Years at Company Analysis

Employee tenure can influence attrition patterns.

This analysis examines whether employees with fewer years at the organization are more likely to leave.

```python
# ============================================
# Years at Company Distribution
# ============================================

plt.figure(figsize=(8,5))

sns.histplot(data=df, x='YearsAtCompany', hue='Attrition', kde=True)

plt.title("Years at Company vs Attrition")
plt.xlabel("Years at Company")
plt.ylabel("Frequency")

plt.show()
```

### Insight

Attrition patterns across employee tenure help identify whether:
- New employees leave early
- Long-term employees face disengagement
- Retention challenges exist at specific career stages

# Correlation Analysis

Correlation analysis helps identify relationships between numerical variables.

It is useful for:
- Feature selection
- Detecting multicollinearity
- Understanding influential variables

```python
# ============================================
# Correlation Heatmap
# ============================================

plt.figure(figsize=(16,10))

correlation_matrix = df.corr(numeric_only=True)

sns.heatmap(correlation_matrix, cmap='coolwarm')

plt.title("Feature Correlation Heatmap")

plt.show()
```

### Insight

The correlation heatmap highlights relationships among numerical variables.

Strong positive or negative correlations may indicate:
- Important predictive features
- Redundant variables
- Potential drivers of employee attrition
