```python
# ============================================
# Import Required Libraries
# ============================================

import pandas as pd
import numpy as np

import shap
import joblib

import matplotlib.pyplot as plt

# Display settings
pd.set_option('display.max_columns', None)

print("Libraries imported successfully.")
```

# Load Dataset and Trained Model

In this section:
- The processed dataset is loaded
- The trained machine learning model is restored
- Features and target variables are prepared for explainability analysis

```python
# ============================================
# Load Dataset
# ============================================

df = pd.read_csv("../data/final_employee_attrition.csv")

# ============================================
# Load Trained Model
# ============================================

model = joblib.load("../models/best_model.pkl")

print("Dataset and trained model loaded successfully.")
```

```python
# Separate features and target variable

X = df.drop('Attrition', axis=1)

y = df['Attrition']

print("Feature preparation completed.")
```

# Introduction to SHAP

SHAP (SHapley Additive exPlanations) is an explainable AI technique that helps interpret machine learning predictions.

SHAP provides:
- Global model interpretability
- Feature importance analysis
- Individual prediction explanations
- Human-understandable AI insights

This is especially important in HR analytics because:
- HR leaders require transparent predictions
- Retention decisions should be explainable
- Workforce intelligence systems must build organizational trust

# SHAP Explainer Initialization

The SHAP explainer is initialized to interpret model behavior and calculate feature contribution values.

```python
# ============================================
# Initialize SHAP Explainer
# ============================================

explainer = shap.TreeExplainer(model)

shap_values = explainer.shap_values(X)

print("SHAP values generated successfully.")
```

# Global Feature Importance Analysis

Global explainability identifies:
- The most influential features
- Overall drivers of employee attrition
- Variables contributing most strongly to model predictions

This helps HR teams understand organizational attrition patterns at scale.

```python
# ============================================
# SHAP Summary Plot
# ============================================

shap.summary_plot(shap_values, X)
```

### Insight

The SHAP summary plot highlights the features with the strongest influence on employee attrition predictions.

Features appearing at the top contribute most significantly to the model’s decision-making process.

Typical high-impact drivers may include:
- Overtime
- Job satisfaction
- Monthly income
- Work-life balance
- Promotion delays

# Feature Importance Ranking

This visualization ranks features according to their average impact on attrition prediction outcomes.

```python
# ============================================
# SHAP Feature Importance Bar Plot
# ============================================

shap.summary_plot(
    shap_values,
    X,
    plot_type="bar"
)
```

### Insight

The feature importance ranking provides a high-level overview of the most influential workforce factors associated with attrition risk.

These insights can support:
- Retention strategy planning
- HR policy improvements
- Workforce engagement initiatives

# Individual Employee Risk Explanation

SHAP can also explain predictions at the employee level.

This enables HR teams to understand:
- Why a specific employee is considered high-risk
- Which factors contribute most strongly to attrition probability

```python
# ============================================
# Select Sample Employee
# ============================================

sample_index = 10

sample_employee = X.iloc[sample_index]

print("Sample employee selected successfully.")
```

```python
# ============================================
# SHAP Force Plot
# ============================================

shap.initjs()

shap.force_plot(
    explainer.expected_value,
    shap_values[sample_index],
    sample_employee
)
```

### Insight

The force plot explains how individual features influence a specific employee’s attrition prediction.

Features pushing prediction higher increase attrition risk, while features pushing prediction lower reduce the risk score.

This level of explainability enables:
- Transparent HR decision-making
- Personalized retention interventions
- Better workforce intelligence
