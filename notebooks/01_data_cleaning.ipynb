# Missing Value Analysis

Missing values can negatively impact machine learning performance.

This section identifies:
- Columns containing null values
- Number of missing records per feature

```python
# Check missing values in each column
missing_values = df.isnull().sum()

missing_values
```

# Duplicate Record Analysis

Duplicate records may introduce bias into the model and affect prediction performance.

This section checks whether duplicate employee records exist in the dataset.

```python
# Check duplicate records
duplicate_count = df.duplicated().sum()

print(f"Duplicate Records Found : {duplicate_count}")
```

```python
# Remove duplicate rows if present

if duplicate_count > 0:
    df.drop_duplicates(inplace=True)
    print("Duplicate records removed successfully.")
else:
    print("No duplicate records found.")
```

# Target Variable Distribution

The target variable is:
- Attrition
    - 1 → Employee left the organization
    - 0 → Employee stayed

Understanding class distribution is important because:
- Imbalanced datasets may bias machine learning models
- Additional techniques like SMOTE may be required

```python
# Display attrition class distribution
df['Attrition'].value_counts()
```

# Observation

The attrition distribution helps determine whether the dataset is balanced or imbalanced.

If one class significantly dominates the other:
- The model may become biased
- Additional balancing techniques such as SMOTE may be necessary during model training

# Feature Type Identification

Separating categorical and numerical columns helps simplify:
- Encoding
- Scaling
- Feature engineering
- Model preprocessing

```python
# Identify categorical columns
categorical_columns = df.select_dtypes(include='object').columns.tolist()

# Identify numerical columns
numerical_columns = df.select_dtypes(exclude='object').columns.tolist()

print("Categorical Columns:")
print(categorical_columns)

print("\nNumerical Columns:")
print(numerical_columns)
```
