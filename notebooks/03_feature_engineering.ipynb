```python
# ============================================
# Import Required Libraries
# ============================================

import pandas as pd
import numpy as np

from sklearn.preprocessing import LabelEncoder

# Display settings
pd.set_option('display.max_columns', None)

print("Libraries imported successfully.")
```

# Load Cleaned Dataset

The cleaned employee attrition dataset is loaded for feature engineering and preprocessing.

```python
# ============================================
# Load Cleaned Dataset
# ============================================

df = pd.read_csv("../data/cleaned_employee_attrition.csv")

print("Dataset loaded successfully.")
```

```python
# Display first 5 rows
df.head()
```

# Feature Engineering

Feature engineering improves model performance by creating meaningful variables from existing data.

In this section, additional workforce intelligence features are created to better capture employee behavior and attrition risk patterns.

## Income-to-Experience Ratio

This feature measures employee compensation relative to total professional experience.

### Purpose
- Identify employees who may feel underpaid
- Capture compensation imbalance
- Improve attrition risk prediction

```python
# ============================================
# Create Income-to-Experience Ratio
# ============================================

df['IncomeExperienceRatio'] = (
    df['MonthlyIncome'] / (df['TotalWorkingYears'] + 1)
)

print("IncomeExperienceRatio feature created successfully.")
```

## Promotion Delay Indicator

Employees experiencing long promotion delays may become disengaged and more likely to leave the organization.

### Logic
- 1 → Promotion delay exists
- 0 → No major promotion delay

```python
# ============================================
# Create Promotion Delay Feature
# ============================================

df['PromotionDelay'] = np.where(
    df['YearsSinceLastPromotion'] > 3,
    1,
    0
)

print("PromotionDelay feature created successfully.")
```

## Employee Engagement Score

Employee engagement is estimated using:
- Job Satisfaction
- Environment Satisfaction
- Relationship Satisfaction
- Work-Life Balance

Higher engagement scores generally indicate:
- Better employee experience
- Lower attrition probability

```python
# ============================================
# Create Employee Engagement Score
# ============================================

df['EngagementScore'] = (
    df['JobSatisfaction'] +
    df['EnvironmentSatisfaction'] +
    df['RelationshipSatisfaction'] +
    df['WorkLifeBalance']
)

print("EngagementScore feature created successfully.")
```

## Workload Stress Indicator

Employees with excessive overtime and frequent travel may experience higher workplace stress.

This feature identifies employees potentially exposed to elevated workload pressure.

```python
# ============================================
# Create Workload Stress Flag
# ============================================

df['WorkStressFlag'] = np.where(
    (df['OverTime'] == 'Yes') &
    (df['BusinessTravel'] == 'Travel_Frequently'),
    1,
    0
)

print("WorkStressFlag feature created successfully.")
```

# Engineered Feature Preview

Displaying the newly created features for validation and inspection.

```python
# Display engineered features

df[[
    'IncomeExperienceRatio',
    'PromotionDelay',
    'EngagementScore',
    'WorkStressFlag'
]].head()
```

# Categorical Variable Encoding

Machine learning algorithms require numerical input.

This section converts categorical features into numerical representations using Label Encoding.

```python
# ============================================
# Identify Categorical Columns
# ============================================

categorical_columns = df.select_dtypes(include='object').columns

print("Categorical Columns:")
print(categorical_columns)
```

# Label Encoding

Each categorical feature is transformed into numerical labels to prepare the dataset for machine learning model training.

```python
# ============================================
# Apply Label Encoding
# ============================================

label_encoder = LabelEncoder()

for column in categorical_columns:
    df[column] = label_encoder.fit_transform(df[column])

print("Categorical encoding completed successfully.")
```
